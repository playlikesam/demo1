Evaluating a Model

Once the model is trained, it needs to be tested to see if it would operate well in the real world.

Hyper Parameter Tuning

Hyper parameter is a parameter whose value is set before the learning process begins. This step helps to improve the performance of the model.

Prediction

Prediction refers to the output of the algorithm after it has been trained on past dataset. The output will generate values for unknown input variables and the model accuracy is thus tested.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 1)
```
